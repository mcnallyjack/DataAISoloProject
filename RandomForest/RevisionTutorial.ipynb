### Introduction <a class="anchor" id="IntroDecTree"><a/>

The Decision Tree machine learning algorithm is a predictive approach used in Machine Learning. In simple terms a decision tree takes a large dataset and divides it into progressively smaller datasets using a particular set of decision rules [1].  

 

Decision Trees or DTâ€™s as they are sometimes referred to are a non-parametric model. This means that you may not know the normal distribution of the dataset. Decision Trees can be used for both Regression and Classification models which is rather helpful because classification is about predicting a label while regression is about predicting a quantity. The fact that both models can be predicted using decision trees renders it rather powerful in machine learning [2].  

 

Decision trees do require some key values to come to fruition such as [3];  

 

- <b> Attribute Value Description </b> = the object must be expressible in a fixed collection of attributes. E.g. Cold or Warm. <br>
- <b> Sufficient Data</b>  = enough data must be provided so the model can be as accurate as possible. <br>
- <b> Target Values </b>  = the target function has discrete output values.  

 

Decision Trees sometimes use builders. Builders are algorithms that have the ability to generate a tree. Some examples include Random Forrest, Classification and Regression Trees (CART), ID3 and C4.5. One of the most influential algorithms is the CART algorithm and this is the algorithm that will demoed later on in this section. 

 



 <i>Decision Tree [10]</i>

 <h4> Different Decision Tree Algorithms </h4>
 
 - <b>CART Algorithm </b>

The CART algorithm or the Classification and Regression Tree algorithm introduced by Leo Breiman [8] is a non-parametric algorithm that produces classification or regression trees. 

The algorithm uses Gini impurity is used to decide the optimal split of the root node. Gini impurity displays the probability of misclassifying an observation [9]. 

The CART algorithm will be the algorithm that will be demoed for decision trees. The data pre-processing and the main algorithm will be demonstrated using Entropy and Gini-impurity to see which method more suits the dataset. 

- <b>ID3 Algorithm</b> 
 
The ID3 algorithm is an algorithm created by Ross Quinlan and is used to generate a tree from a dataset.  

This Algorithm gets its name from what it actually does, â€œIterative Dichotomiserâ€ or in plainer terms â€œRepetitively Dividingâ€ basically divides the values into two or more groups at each step. 

The main method of decision making the ID3 algorithms uses is â€œEntropyâ€. Entropy is the measure of the amount of uncertainty in a dataset. For example, if the values are always the same e.g. answers to a question is always â€œNoâ€ the entropy is said to be low, but if the values are different, e.g. the answers are â€œYesâ€, â€œNoâ€, â€œUnknownâ€ the entropy is said to be High and this is when it is most effective [4]. 

- <b>C4.5 Algorithm </b>     

Another algorithm that is used to create decision trees is the C4.5 algorithm this algorithm is the precursor to the ID3 algorithm.  

The C4.5 algorithms which was once again created by Ross Quinlan is an improved version of the ID3 algorithm, the C4.5 algorithm offers these key improvements [5]; 

   - C4.5 handles both continuous and discrete data, in order to handle continuous data C4.5 creates a threshold and separates the values into lists that are greater than an less than the thresholds. 
   - C4.5 marks the missing data as â€˜?â€™ so they are not used in gain and entropy calculations. 
   - Lastly C4.5 prunes the tree after the creation. This means that the algorithm goes back through the tree and removes branches that are not necessary and replaces them with leaf nodes. 

 

- <b> Advantages of Decision Trees</b>

    - Decision trees one of the more easier to understand algorithms and most people can understand them after a brief explanation [6]. 

    - The best, worst and expect values can all be predicted using this algorithm [6]. 

    - The decision tree algorithm can be combined with other algorithms [6]. 
    

- <b> Disadvantages of Decision Trees </b>

   -  The decision tree algorithm can be unstable, if the value change even the smallest amount the whole algorithm can be thrown off course [7]. 

   - The calculations associated with decision trees can sometimes be very complex and difficult to understand [7]. 
   - Decision tree training takes more time and is relatively expensive in terms of time and complexity [7].

As can be seen in the cell below the dataset has some null values, we will handle them using an average replacement method, this involves getting the average value of the column and then filling the null values with the average value. This means that Female occurs only '112' times in the dataset, while male appears '488' times. Therefore I will replace the Null values with female.

Now you see that the Gender column has no null values, I will now repeat this for all of the remaining columns that have null values present.

```python
LoanApproval['Married'] = LoanApproval['Married'].fillna("No")
LoanApproval['Dependents'] = LoanApproval['Dependents'].fillna('3')
LoanApproval['Self_Employed'] = LoanApproval['Self_Employed'].fillna("Yes")
LoanApproval['LoanAmount'] = LoanApproval['LoanAmount'].fillna('66.0')
LoanApproval['Loan_Amount_Term'] = LoanApproval['Loan_Amount_Term'].fillna('270.0')
LoanApproval['Credit_History'] = LoanApproval['Credit_History'].fillna('0.0')
```

#### Converting to binary

As you can see in the cell above the dataset now has no null values. I will now move onto converting the string decisons to binary decisions, e.g. Male will become 1 and Female will become 0. This process can be completed using the LabelEncoder function but some columns do not need to be changed, Therefore I will complete this process manually.

```python
LoanApproval['Gender'] = LoanApproval['Gender'].map(dict(Male=1,Female=0))
LoanApproval['Married'] = LoanApproval['Married'].map(dict(Yes=1,No=0))
LoanApproval['Education'] = LoanApproval['Education'].eq('Graduate').mul(1.0)
LoanApproval['Self_Employed'] = LoanApproval['Self_Employed'].eq('Yes').mul(1.0)
LoanApproval['Property_Area'] = LoanApproval['Property_Area'].map(dict(Urban=1,Rural=0,Semiurban=2))
LoanApproval['Dependents'] = LoanApproval['Dependents'].eq('3+').mul(3.0)
LoanApproval['Loan_Status'] = LoanApproval['Loan_Status'].map(dict(Y=1,N=0))
```

### Introduction <a class="anchor" id="IntroRanFor"></a>

The Random forests machine learning algorithm is a supervised machine learning algorithm that is an ensemble of decision of decision trees.[11] The overall goal of Random forests is to combine many decision trees to go get a much better result.

Random forest like decision trees and many other algorithms can be used for Classification and Regression problems. As mentioned above Random forest use a method called "ensemble", this means that the Random forest classifer creates multiple decision trees and then selects the best solution by voting.[12]

Random forests are very similar to Decision trees, Random forests needs to meet the following requirments to be created; 

- <b> Attribute Value Description </b> = the object must be expressible in a fixed collection of attributes. E.g. Cold or Warm. <br>
- <b> Target Values </b>  = the target function has discrete output values.  <br>
- <b> Feature Values</b>  = values thats the target value can be compared against 




#### Bagging
Bagging or (Bootstrap Aggregating) is a method used by Random forests to generate new datasets for a model. Bagging is a very important technique as it helps the Random forest algorithm to prevent overfitting. Overfitting is a big problem in Decision trees.[13]

Bagging also helps the random forest algorithm become much more stable and much more accurate

 <i>Random Forests [14]</i>

- <b> Advantages of Random forests</b>

    - One of the most accurate machine learning algorithms and also works well on large datasets. [15]
    
    - Flexiable to both Classification & Regression problems [16].
    
    - Outputs the importance of Features [17].
    
    
- <b> Disadvantages of Random forests </b>

    - Random forests tend to overfit noisy datasts[15]
    
    - Randoms forests result may change drastically if there is a small change in the data. [17]
    
    - Take longer to train the model. [18]

#### Splitting the Data

```python
features = ['gender', 'age', 'hypertension','heart_disease', 'ever_married','work_type','Residence_type','avg_glucose_level','bmi', 'smoking_status']
X = StrokePred[features]
y = StrokePred.stroke
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

   - <b> X </b> = features of the dataset
   - <b> y </b> = stroke
   
I then pass the X and y variables into the <b> train_test_split </b> function, like will split the variables using the test_size parameter that I set.

#### Training the Model & Accuracy

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 100, random_state = 0, max_depth=3)
rf.fit(X_train, y_train);
y_pred = rf.predict(X_test)
```
